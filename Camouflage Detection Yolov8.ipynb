```python
import os
from ultralytics import YOLO
model = YOLO('yolov8n-seg.pt')
model.train(data='config.yaml', epochs=10, imgsz=256, batch=32)
```

```python
import cv2
import matplotlib.pyplot as plt

model_path = '/content/drive/My Drive/YOLOv8/runs/segment/train/weights/last.pt'
image_path = '/content/drive/My Drive/YOLOv8/data/images/val/image501.jpg'

img = cv2.imread(image_path)
H, W, _ = img.shape

model = YOLO(model_path)

results = model(img)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image')

for result in results:
    for j, mask in enumerate(result.masks.data):
        mask = mask.numpy() * 255
        mask = cv2.resize(mask, (W, H))
        axs[1].imshow(mask, cmap='gray')
        axs[1].set_title('Segmentation Mask')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

ground_truth_mask_path = '/content/drive/My Drive/YOLOv8/image501.png'
ground_truth_mask = cv2.imread(ground_truth_mask_path, cv2.IMREAD_GRAYSCALE)

predicted_mask = mask

intersection = np.logical_and(ground_truth_mask, predicted_mask).sum()
union = np.logical_or(ground_truth_mask, predicted_mask).sum()

iou = intersection / union

print("Intersection over Union (IoU):", iou)
```
